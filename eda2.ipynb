```python
import pandas as pd
import numpy as np 
from ctgan import CTGAN
import matplotlib.pyplot as plt
```

```python
# Reading in the burger king menus
data = pd.read_csv('burger-king-menu.csv')
```

```python
data.head()
```

```python
data["Category"].unique()
```

```python
# Need to preprocess the burgerframe

data = data.drop(labels="Item", axis="columns")
```

```python
discrete_columns = [
    "Category"
]

encode_categorical =  {"Category":     {"Burgers": 1, "Chicken": 2, "Breakfast": 3}
               }
```

```python
#Replace strings with int in category column

data = data.replace(encode_categorical)
```

```python
#Training model
ctgan = CTGAN(epochs=1000)
ctgan.fit(data, discrete_columns)
```

```python
# Creating synthetic data

synthetic_data = ctgan.sample(10)
```
